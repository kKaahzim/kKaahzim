#### b-) Ler o dataset "*Hourly_wages/hourly_wages.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e as 5 últimas linhas do dataset.

```
caminho_arquivo = 'https://raw.githubusercontent.com/malegopc/DSBD/main/Datasets/Hourly_wages/hourly_wages.csv'

df_wages = pd.read_csv(caminho_arquivo)

print("--- 5 Primeiras Linhas ---")
display(df_wages.head())

print("\n--- 5 Últimas Linhas ---")
display(df_wages.tail())
```

#### c-) Obter um resumo da estatística descritiva dos dados utilizando um único comando (função).

```
df_wages.describe()
```

#### d-) Calcule a proporção de funcionários de sexo masculino e feminino.

```
proporcao_sexo = df_wages['female'].value_counts(normalize=True) * 100
print("Proporção por sexo (%):")
print(proporcao_sexo)
```

#### e-) Calcule a proporção de casados e solteiros.

```
proporcao_estado_civil = df_wages['marr'].value_counts(normalize=True) * 100
print("Proporção por estado civil (%):")
print(proporcao_estado_civil)
```

#### f-) Calcule as médias de cada uma das variáveis: anos de educação, anos de experiência e idade

```
medias = df_wages[['education_yrs', 'experience_yrs', 'age']].mean()
print("Médias:")
print(medias)
```

#### g-) Calcule as medianas de cada uma das variáveis: anos de educação, anos de experiência e idade.

```
medianas = df_wages[['education_yrs', 'experience_yrs', 'age']].median()
print("Medianas:")
print(medianas)
```

####h-) Calcule as modas de cada uma das variáveis: anos de educação, anos de experiência e idade.

```
modas = df_wages[['education_yrs', 'experience_yrs', 'age']].mode()
print("Modas:")
display(modas)
```

####i-) Analisando especificamente a variável **idade**, responda: qual é a idade do conjunto de dados em que **70%** dos funcionários apresentam idade inferior?

```
idade_70 = df_wages['age'].quantile(0.70)
print(f"70% dos funcionários têm idade inferior a {idade_70} anos.")
```

####j-) Analisando especificamente a variável **anos de experiência**, responda: qual é a quantidade de anos de experiência no conjunto de dados em que 10% dos funcionários ainda não possuem?

```
exp_90 = df_wages['experience_yrs'].quantile(0.90)
print(f"10% dos funcionários possuem mais de {exp_90} anos de experiência (logo, ainda não alcançaram esse valor inferior).")
```

####k-) Calcule o desvio-padrão de cada um dos atributos: anos de educação, anos de experiência e idade.

```
desvios = df_wages[['education_yrs', 'experience_yrs', 'age']].std()
print("Desvio-Padrão:")
print(desvios)
```

####l-) Determine qual dos três atributos apresenta maior variabilidade: anos de educação, anos de experiência ou idade? Justifique a sua resposta.

```
cv = (df_wages[['education_yrs', 'experience_yrs', 'age']].std() / df_wages[['education_yrs', 'experience_yrs', 'age']].mean()) * 100
print("Coeficiente de Variação (%):")
print(cv)
atributo_maior_var = cv.idxmax()
print(f"\nO atributo com maior variabilidade é '{atributo_maior_var}' porque apresenta o maior Coeficiente de Variação.")
```

####m-) Leia o conjunto de dados [Atlas_Brasil_2014.csv](https://github.com/malegopc/DSBD/blob/main/Datasets/Atlas/Atlas_Brasil_2014.csv)

```
caminho_atlas = 'https://raw.githubusercontent.com/malegopc/DSBD/main/Datasets/Atlas/Atlas_Brasil_2014.csv'
df_atlas = pd.read_csv(caminho_atlas)
display(df_atlas.head())
```

n-) Calcule a média do IDHM.

```
media_idhm = df_atlas['IDHM'].mean()
print(f"Média do IDHM: {media_idhm}")
```

o-) Qual é a taxa de mortalidade em que 50% dos estados apresentam taxa inferior?

```
mediana_mortalidade = df_atlas['MORT1'].median()
print(f"A taxa de mortalidade em que 50% dos estados apresentam taxa inferior é: {mediana_mortalidade}")
```

p-) Calcule qual o valor de RDPC (renda per capita) em que 25% dos estados apresentam valor superior.

```
rdpc_75 = df_atlas['RDPC'].quantile(0.75)
print(f"25% dos estados apresentam uma Renda Per Capita (RDPC) superior a: {rdpc_75}")
```

q-) Determine dentre as duas variáveis: T_ANALF25M (Taxa de analfabetismo - 25 anos ou mais) ou RDPC (Renda per capita média), qual apresenta maior variabilidade. Justifique a sua resposta.

```
cols = ['T_ANALF25M', 'RDPC']
cv_atlas = (df_atlas[cols].std() / df_atlas[cols].mean()) * 100
print("Coeficiente de Variação (%):")
print(cv_atlas)

maior_var_atlas = cv_atlas.idxmax()
print(f"\nA variável com maior variabilidade é '{maior_var_atlas}', pois possui o maior Coeficiente de Variação, "
      "o que é a medida estatística correta para comparar a dispersão relativa de variáveis com naturezas e escalas numéricas distintas.")
```
